#### Question 2 : Tableaux des noms de villes à visiter

````python
def noms_villes(villes):
    """
    Cette fonction permet de créer un tableau contenant le nom des villes
    
    :param list villes: Tableau de villes contenant le nom des villes, leur longitude et leur latitude.
    :returns: nomvilles, un tableau contenant le nom des villes 
    """
    nomvilles = [] # On crée un tableau vide nomvilles
    for i in range(len(villes)): # On parcours le tableau
        if type(villes[i]) == str: # On regarde si l'élément d'indice i du tableau villes est une chaine de caractère
            nomvilles.append(villes[i]) # On ajoute au tableau nomvilles le nom de la ville
    return nomvilles

def test_noms_villes():
    """
    Cette fonction permet de vérifier si la fonction nom_villes est valide ou non
    """
    assert noms_villes(villes) == ["Paris","Lyon","Marseille","Lille","Strasbourg","Rennes","Clermont-Ferrand","Bordeaux","Toulouse","Grenoble"]
    assert not noms_villes(villes) == ["Villetaneuse","Paris","Lyon","Marseille","Lille","Strasbourg","Rennes","Clermont-Ferrand","Bordeaux","Toulouse","Grenoble"]
    assert not noms_villes(villes) == ["Lyon","Marseille","Lille","Strasbourg","Rennes","Clermont-Ferrand","Bordeaux","Toulouse","Grenoble"]
    print("Test de la fonction noms_villes : ok")
````

#### Question 3 : Distance entre deux villes repérées par leurs longitude et latitude

Pour calculer la distance $d$ entre deux villes $V_1$ (de longitude $long1$ et de latitude $lat1$) et $V_2$ (de longitude $long2$ et de latitude $lat2$) à partir de leurs longitude et latitude, on peut utiliser la formule suivante :

$d=R*arccos(sin(lat1)*sin(lat2)+cos(lat1)*cos(lat2)*cos(long2-long1))$

où le rayon de la terre $R$ vaut 6370.7km et les latitudes et longitudes sont exprimées en radians (en non en degrés décimaux).
Cela correspond à la méthode de calcul avec la loi des sinus décrite [ici](http://villemin.gerard.free.fr/aGeograp/Distance.htm) 

Par exemple, Paris et Marseille sont distantes de 661,86 km.

- Ecrire une fonction `distance(long1,lat1,long2,lat2)` qui calcule et retourne la distance entre deux villes $V_1$ (de longitude `long1` et `lat1`) et $V_2$ (de `long2`et `lat2`).

**NB** Les fonctions `sin`, `cos` et le nombre `pi` sont définis dans la bibliothèque `math`. La fonction `arcos` est définie dans la bibliothèque `numpy`.

Pour **convertir des degrés en radians**, il faut multiplier par $\pi/180$.


- Ecrire une fonction de tests unitaire permettant de vérifier cette fonction.

````python
from math import *
from numpy import *

def distance(long1,lat1,long2,lat2):
    """
    Cette fonction permet de calculer la distance entre deux villes via leurposition en longitude et en latitude.
    
    :param float long1: Longitude de la première ville.
    :param float lat1: Latitude de la première ville.
    :param float long2: Longitude de la deuxième ville.
    :param float lat2: Latitude de la deuxième ville.
    :returns: Résultat de la formule permettant de calculer la distance entre deux villes à partir de leur 
              longitude et de leur latitude.
    """
    # Début des conversions en radian
    radlong1 = long1*(pi/180) 
    radlat1 = lat1*(pi/180)
    radlong2 = long2*(pi/180)
    radlat2 = lat2*(pi/180)
    # Fin des conversions
    return 6370.70*acos(sin(radlat1)*sin(radlat2) + cos(radlat1)*cos(radlat2)*cos(radlong2-radlong1)) # Formule permettant de calculer la distance entre deux villes

def test_distance():
    """
    Cette fonction permet de vérifier si la fonction distance est valide ou non
    """
    # isclose permet de vérifier si un flottant est environ égal à un autre flottant
    assert isclose(661.86, distance(2.33, 48.86, 5.40, 43.30)) # La distance entre Paris et Marseille est de 661.86 km
    assert isclose(443.92, distance(3.08, 45.77, -1.66, 48.11)) # La distance entre Clermont-Ferrant et Rennes est de 443.92km
    print("Test de la fonction distance : ok")
````

#### Question 4 : Coordonnées d'une ville dans le tableau `villes`

a) Ecrire une fonction `indexCity(ville, villes)` qui retourne la position (l'indice) de la `ville` passée en paramètre dans le tableau `villes`. Elle retournera `-1` si la ville passée en paramètre n'est pas dans le tableau. Cette information pourra être utilisée dans la suite pour trouver la longitude et la latitude d'une ville donnée par son nom.

Par exemple, 
- l'instruction `indexCity("Paris", villes)` retournera `0`,
- l'instruction `indexCity("Clermont-Ferrand",villes)` retournera  `18`,
- et l'instruction `indexCity("Villetaneuse",villes)` retournera  `-1`.

b) Ecrire une fonction de tests unitaire permettant de vérifier cette fonction.

````python
def indexCity(ville, villes):
    """
    Cette fonction permet de connaitre l'indice d'une ville si elle est dans un tableau ville.
    
    :param str ville: Nom de la ville dont on recherche l'indice.
    :param list villes: Tableau de villes contenant le nom des villes, leur longitude et leur latitude.
    :returns: Indice de la ville passé en paramètre si elle est dans le tableau ville, -1 sinon.
    """
    for i in range(0, len(villes), 3): # On parcours le tableau villes par indice
        if villes[i] == ville:
            return i
    return -1

def test_indexCity():
    """
    Cette fonction permet de vérifier si la fonction test_indexCity est valide ou non 
    """
    assert indexCity("Paris", villes) == 0
    assert indexCity("Clermont-Ferrand", villes) == 18
    assert indexCity("Villetaneuse", villes) == -1
    print("Test de la fonction indexCity : ok")
````

#### Question 5 : Distance entre deux villes données par leurs noms

- Ecrire une fonction `distance_noms(nom1,nom2, villes)` qui calcule et retourne la distance entre deux villes $V_1$ et $V_2$ appelées `nom1` et `nom2` dans le tableau `villes`. La fonction retournera `-1` si une de ces villes n'est pas dans le tableau.
- Ecrire une fonction de tests unitaires permettant de vérifier cette fonction.

````python
def distance_noms(nom1, nom2, villes):
    """
    Cette fonction permet de connaitre la distance entre deux villes via leurs noms.
    :param str nom1: Nom de la première ville.
    :param str nom2: Nom de la deuxième ville.
    :param list villes: Tableau de villes contenant le nom des villes, leur longitude et leur latitude.
    :returns: Distance entre les deux villes si elles sont dans le tableau villes, et -1 sinon. 
    """
    if nom1 not in villes or nom2 not in villes:
        return -1
    else:
        indville1 = indexCity(nom1, villes)
        indville2 = indexCity(nom2, villes)
        return distance(villes[indville1+1],villes[indville1+2],villes[indville2+1],villes[indville2+2])
    
def test_distance_noms():
    """
    Cette fonction permet de vérifier si la fonction distance_noms est valide ou non 
    """
    assert isclose(661.86, distance_noms("Paris","Marseille",villes))
    assert distance_noms("Paris","Villetaneuse",villes) == -1
    print("Test de la fonction distance_noms : ok")
````

#### Question 6 :  Lecture des villes d'une tournée à partir d'un fichier
On suppose que les données sur les villes où doit se rendre le Père Noël réseau sont stockées dans un fichier CSV de la manière suivante :
```
ville1;longitude1;latitude1
ville2;longitude2;latitude2
ville3;longitude3;latitude3
...
```
Autrement dit, chaque ligne du fichier contient le nom d'une ville, sa longitude et sa latitude en degrés décimaux, ces données étant séparées par des `';'`. Ces villes sont celles que le Père Noël doit desservir

**NB** : Six fichiers CSV de ce type sont fournis dans le répertoire `files/`, il s'agit des fichiers `africa.csv`, `asia.csv`,`europe.csv`, `north-america`, `oceania.csv` and `south-america.csv`. 
Il est recommandé d'en fabriquer d'autres. 

- Ecrire une fonction `lecture_villes(path)` prenant en paramètre un chemin vers un tel fichier CSV et retournant un tableau contenant les informations sur les villes à visiter.

````python
def lecture_villes(path):
    """
    Cette fonction permet d'obtenir un tableau de villes qui respecte le format du tableau villes.
    :param str path: Un chemin vers un fichier CSV.
    :returns: Un tableau contenant les informations sur les villes à visiter.
    """
    tabville = [] # On crée un tableau vide tabville
    f = open(path, "r", encoding = "utf-8") # On ouvre le fichier en lecture
    li = f.readlines() # On lit toutes les lignes et on les stocke dans un tableau li
    f.close() # On ferme le fichier
    for elt in li: # On parcours les éléments du tableau li
        ligne = elt.split(";") # On sépare la ligne sous forme d'un tableau en utilisant le ";" comme séparateur
        tabville.append(ligne[0]) # On ajoute le premier élément de ligne au tableau tabville, donc le nom de la ville
        tabville.append(float(ligne[1])) # On ajoute le deuxième élément de ligne au tableau tabville en le castant en flottant, donc la longitude de la ville
        tabville.append(float(ligne[2].strip())) # On ajoute le troisième élément de ligne au tableau tabville en enlevant le "\n" grâce à la fonction strip() et en le castant en flottant, donc la latitude de la ville
    return tabville
````

#### Question 7 : Longueur d'une tournée

On considère que le Père Noël peut partir de n'importe laquelle des villes qu'il doit visiter et qu'il doit revenir à son point de départ à la fin de la tournée. Lors de sa tournée, il doit visiter chacune des villes dont les informations sont contenues dans le tableau `villes` une fois et une seule. 

On représente une tournée possible sous la forme d'un tableau `tournee` contenant le nom des villes dans l'ordre où le Père Noël les visitera sans oublier de revenir à la ville dont il est parti. 

L'objectif est de calculer la longueur d'une tournée. 

- Ecrire une fonction `long_tour(villes, tournee)` qui prend en paramètres le tableau `villes` contenant les informations sur les villes à visiter et le tableau `tournee` indiquant l'ordre dans lequel les villes sont visitées. La fonction retourne la longueur de la tournée correspondante.

- Ecrire une fonction de tests unitaires permettant de vérifier cette fonction.

````python
def long_tour(villes, tournee):
    """
    Cette fonction permet de de calculer la longueur d'une tournée  
    :param list villes: Tableau de villes contenant le nom des villes, leur longitude et leur latitude.
    :param list tournee: Tableau contenant les villes à visiter lors de la tournée.
    :returns: La distance totale de la tournee en revenant à la ville de départ.
    """
    distance_totale = 0 
    for i in range(len(tournee)-1): # On parcours le tableau tournee sans compter le dernier élément
        distance_totale += distance_noms(tournee[i], tournee[i+1], villes)# On calcule la ditance entre la ville d'indice i et et la ville d'indice i+1 et on l'ajoute à la distance_totale
    distance_totale += distance_noms(tournee[-1], tournee[0], villes)# On oublie pas de revenir à la ville de départ donc on calcule la distance entre la dernière villes du tableau et la première et on l'ajoute à la distance_totale 
    return distance_totale

#print(long_tour(villes, ["Marseille","Paris","Toulouse","Lyon","Rennes"]))

def test_long_tour():
    assert isclose(long_tour(villes, ["Marseille","Paris","Toulouse","Lyon","Rennes"]),2936.30)
    assert not isclose(long_tour(villes, ["Toulouse","Marseille","Rennes","Paris","Lyon"]),2936.30)
    print("Test de la fonction long_tour : ok")
````

#### Question 8 : Recherche d'une plus courte tournée 

Pour améliorer la longueur de la tournée, on peut essayer de changer l'ordre des villes aléatoirement. Pour une tournée donnée par un tableau `tournee` cela peut se faire avec la fonction `shuffle` de la bibliothèque `random` qui mélange aléatoirement les villes qui sont dans le tableau `tournee`, grâce à l'instruction  `shuffle(tournee)`, et ne retourne rien.

- Utiliser cette fonction pour écrire un programme qui cherche une tournée plus courte à proposer au Père Noël. 
- Comment la solution s'améliore-t-elle en fonction de nombre de tentatives de mélange ? 

````python
from random import shuffle

liste_villes = lecture_villes(input("Veuillez entrer un fichier .csv de villes : "))
tournee_aleatoire = noms_villes(liste_villes) 
tournee_la_plus_courte = None # Nous n'avons pas de tournee la plus courte au début, on crée simplement la variable sans affectation.
dist_plus_courte = 999999999999 # On met une distance extrêmement grande afin que la boucle fonctionne sans condition particulière. 

for i in range (100000): # Plus on fait d'essai, plus la boucle est longue et plus le résultat sera précis.
    shuffle(tournee_aleatoire) # On mélange l'ordre de la tournée
    if long_tour(liste_villes, tournee_aleatoire) < dist_plus_courte: # Si notre nouvelle tournee est plus courte que celle gardé
        dist_plus_courte = long_tour(liste_villes, tournee_aleatoire) # On modifie la distance la plus courte par celle de la nouvelle tournée
        tournee_la_plus_courte = tournee_aleatoire.copy() # On copie le tableau afin d'éviter que ces deux variables renvoient au même espace mémoire
````
